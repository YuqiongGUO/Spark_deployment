```python
from sklearn.datasets import make_blobs
from matplotlib import pyplot
import pandas as pd

data,target=make_blobs(n_samples=2000000,n_features=4,centers=3,cluster_std=[1.0,3.0,2.0])

#在2D图中绘制样本，每个样本颜色不同
pyplot.scatter(data[:,0],data[:,1],c=target);
pyplot.show()
```
